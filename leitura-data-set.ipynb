```python
'''
Cria um array contendo as informações de danceability e energy
'''
X = df.iloc[:,6:8].values
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Normaliza os dados dos array X
standarlization = StandardScaler()
Xst = standarlization.fit_transform(X)
```

```python
Xst[:5]
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
wcss = []

for i in range(1,10):
    kmeans = KMeans(n_clusters=i, init='random')
    kmeans.fit(Xst)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,10),wcss)
plt.title('O Metodo Elbow')
plt.xlabel('numero de Clusters')
plt.ylabel('WSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
labels = kmeans.labels_
labels
```

```python
#PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(Xst)
print("Variância explicada de cada dimensão\n",pca.explained_variance_ratio_)
```

```python
plt.scatter(X_train_pca[:, 0], X_train_pca[:,1], s = 100, c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'red',label = 'Centroids')
plt.title('Spotify Musics Clusters and Centroids')
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.legend()
plt.gcf().set_size_inches((15,12))
plt.show()
```

```python
y_pred = KMeans(n_clusters=4).fit_predict(Xst)
df["cluster"] = y_pred
df.loc[ :, ["country", "cluster", "popularity", "danceability", "energy","loudness"]]
```
