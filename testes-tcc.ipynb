```python
from pymongo import MongoClient
import pandas as pd
import requests
import json
```

```python
def getToken():
    
    url = "https://accounts.spotify.com/api/token"

    payload = "grant_type=client_credentials&undefined="
    headers = {
        'Content-Type': "application/x-www-form-urlencoded",
        'Authorization': "Basic NmRhY2QwYTRhMTE3NDVkNTg2ODY1MTc4MWYzOTBlYjk6MmUwMDY3OWJkY2ZjNDhjOGEwYWQ4Y2M2YWMxYTNiMWY=",
        'cache-control': "no-cache",
        'Postman-Token': "26d5613d-6954-4c55-b39a-b9387316f347"
        }

    response = requests.request("POST", url, data=payload, headers=headers)

    fileJson = json.loads(response.text)

    return fileJson['access_token']
```

```python
'''
    Conecto ao mongo localmente, e acesso o banco 'teste_tcc'
    EntÃ£o, retorno todas as playlists encontradas no documento 'top50'
'''
conect = MongoClient('localhost', 27017)
mongo = conect['teste_tcc']
cursor = mongo['top50'].find()
```

```python
'''
    Transformo meu documento mongo em lista, logo apÃ³s em dataset
'''
df = pd.DataFrame(list(cursor))
```

```python
df = df.drop(columns=['_id', 'collaborative', 'images', 'owner',
                      'primary_color', 'public', 'snapshot_id', 
                      'type','uri'])
df = df.rename(columns={
    "external_urls": "link_spotify",
    "href": "href_playlist",
    "id": "id_playlist",
    "name": "nome",
    "tracks": "href_faixas"
})


df.head()
```

```python
df['num_faixas'] = df['href_faixas']
```

```python
df.head()
```

```python
for index, link in enumerate(df['link_spotify'], start=0):
    df['link_spotify'][index] = link['spotify']
    
for index, link in enumerate(df['href_faixas'], start=0):
    df['href_faixas'][index] = link['href']
    df['num_faixas'][index] = link['total']

df.head()
```

```python
url = df.href_faixas[0]

headers = {
    'Authorization': "Bearer " + getToken()
}

response = requests.request("GET", url, headers=headers)
fileJson = json.loads(response.text)
```

```python
df_faixas_itens = pd.DataFrame(list(fileJson['items']))
```

```python
payload = ""
headers = {
    'Authorization': "Bearer " + getToken()
}

keys = [
    'danceability',
    'energy',
    'loudness', 
    'speechiness',
    'acousticness',
    'instrumentalness',
    'liveness',
    'valence',
    'tempo'
]

# Cria um dicionÃ¡rio para inserir as chaves que desejo
infor = {}
infor = infor.fromkeys(keys)

for key in infor.keys():
    infor[key] = []


url = "https://api.spotify.com/v1/audio-features/" + df_faixas['id'][1]
response = requests.request("GET", url, data=payload, headers=headers)
track_info = json.loads(response.text)
print(track_info)

for index in range(0, len(df_faixas.index)):
    id = df_faixas['id'][index]
    url = "https://api.spotify.com/v1/audio-features/" + id
    response = requests.request("GET", url, data=payload, headers=headers)
    track_info = json.loads(response.text)
    for key in infor.keys():
        infor[key].append(track_info[key])
```
